## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Fill or drop missing values as appropriate
# Drop duplicate rows
titles.drop_duplicates(inplace=True)
credits.drop_duplicates(inplace=True)

# Handle missing values in titles DataFrame

# Fill missing 'description' values with a placeholder
titles['description'].fillna('No description available', inplace=True)

# Fill missing 'age_certification' values with 'Unrated'
titles['age_certification'].fillna('Unrated', inplace=True)

# For 'seasons', fill missing values with 0, assuming movies or shows with no seasons listed have 0 seasons
titles['seasons'].fillna(0, inplace=True)

# For IMDb and TMDB related columns, filling with the mean or median might not be appropriate as it could skew the data.
# For now, we will leave them as is or consider dropping rows/columns later if necessary for specific analyses.
# Dropping rows with missing imdb_id as it's a unique identifier
titles.dropna(subset=['imdb_id'], inplace=True)

# Display the number of remaining missing values after initial handling
print("Missing values in Titles DataFrame after handling:")
print(titles.isnull().sum())
print("\nMissing values in Credits DataFrame after handling:")
print(credits.isnull().sum())
````

#### Chart - 13: Analyze the top actors and directors
